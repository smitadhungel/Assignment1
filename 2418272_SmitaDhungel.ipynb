```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('C:\\Users\\Lenovo\\Downloads\\WHR-2024-5CS037.csv')

# Clean column names by stripping any leading/trailing spaces
data.columns = data.columns.str.strip()

# Filter the data for South Asia countries
south_asia_countries = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka']

# Filter the data to include only South Asian countries
south_asia_df = data[data['Country name'].isin(south_asia_countries)]

# Calculate IQR for 'score' and 'Log GDP per capita'
Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Define the outlier thresholds
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify the outliers for both 'score' and 'Log GDP per capita'
outliers_score = south_asia_df[(south_asia_df['score'] < lower_bound_score) | (south_asia_df['score'] > upper_bound_score)]
outliers_gdp = south_asia_df[(south_asia_df['Log GDP per capita'] < lower_bound_gdp) | (south_asia_df['Log GDP per capita'] > upper_bound_gdp)]

# Display outliers based on Score and GDP per Capita
print("Outlier countries based on Happiness Score:")
print(outliers_score[['Country name', 'score']])

print("\nOutlier countries based on GDP per Capita:")
print(outliers_gdp[['Country name', 'Log GDP per capita']])

# Plotting to visualize outliers
plt.figure(figsize=(10, 6))

# Plotting the Happiness Score distribution
plt.subplot(1, 2, 1)
plt.boxplot(south_asia_df['score'], vert=False)
plt.title('Happiness Score Distribution')
plt.xlabel('Happiness Score')

# Plotting the GDP per Capita distribution
plt.subplot(1, 2, 2)
plt.boxplot(south_asia_df['Log GDP per capita'], vert=False)
plt.title('GDP per Capita Distribution')
plt.xlabel('Log GDP per capita')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('C:\\Users\\Lenovo\\Downloads\\WHR-2024-5CS037.csv')

# Clean column names
data.columns = data.columns.str.strip()

# Calculate IQR for GDP per Capita
gdp_q1 = data['Log GDP per capita'].quantile(0.25)
gdp_q3 = data['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1

# Calculate IQR for Score
score_q1 = data['score'].quantile(0.25)
score_q3 = data['score'].quantile(0.75)
score_iqr = score_q3 - score_q1

# Define outlier thresholds
gdp_lower_bound = gdp_q1 - 1.5 * gdp_iqr
gdp_upper_bound = gdp_q3 + 1.5 * gdp_iqr

score_lower_bound = score_q1 - 1.5 * score_iqr
score_upper_bound = score_q3 + 1.5 * score_iqr

# Identify outliers
data['Outlier'] = ((data['Log GDP per capita'] < gdp_lower_bound) | 
                   (data['Log GDP per capita'] > gdp_upper_bound) | 
                   (data['score'] < score_lower_bound) | 
                   (data['score'] > score_upper_bound))

# Scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='Log GDP per capita', y='score', hue='Outlier', 
                palette={True: 'red', False: 'blue'}, alpha=0.7)

# Add labels and title
plt.title('Scatter Plot of GDP per Capita vs. Score with Outliers Highlighted', fontsize=16)
plt.xlabel('Log GDP per Capita', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.legend(title='Outlier', loc='upper left')
plt.grid(alpha=0.3)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('C:\\Users\Smita\Downloads\\WHR-2024-5CS037.csv')

# Clean column names
data.columns = data.columns.str.strip()

# Define the list of Middle Eastern countries
middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait", "Lebanon",
    "Oman", "Palestine", "Qatar", "Saudi Arabia", "Syria", 
    "United Arab Emirates", "Yemen"
]

# Filter the dataset for Middle Eastern countries
middle_east_df = data[data['Country name'].isin(middle_east_countries)].copy()

# Calculate IQR for 'Log GDP per capita'
gdp_q1 = middle_east_df['Log GDP per capita'].quantile(0.25)
gdp_q3 = middle_east_df['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1

gdp_lower_bound = gdp_q1 - 1.5 * gdp_iqr
gdp_upper_bound = gdp_q3 + 1.5 * gdp_iqr

# Calculate IQR for 'Score'
score_q1 = middle_east_df['score'].quantile(0.25)
score_q3 = middle_east_df['score'].quantile(0.75)
score_iqr = score_q3 - score_q1

score_lower_bound = score_q1 - 1.5 * score_iqr
score_upper_bound = score_q3 + 1.5 * score_iqr

# Identify outliers using .loc to avoid the SettingWithCopyWarning
middle_east_df.loc[:, 'Outlier'] = (
    (middle_east_df['Log GDP per capita'] < gdp_lower_bound) | 
    (middle_east_df['Log GDP per capita'] > gdp_upper_bound) |
    (middle_east_df['score'] < score_lower_bound) | 
    (middle_east_df['score'] > score_upper_bound)
)

# Visualize the outliers using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Log GDP per capita', 
    y='score', 
    data=middle_east_df, 
    hue='Outlier', 
    palette={True: 'red', False: 'blue'}
)

# Add labels and title
plt.title('Middle East Countries: GDP per Capita vs. Score (Highlighting Outliers)', fontsize=14)
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)

# Show the plot
plt.legend(title='Outlier', loc='upper left')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'C:\\Users\Smita\Downloads\\WHR-2024-5CS037.csv'
data = pd.read_csv(file_path)

# List of South Asian countries
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", 
                         "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Filter data for South Asian countries
south_asia_df = data[data['Country name'].isin(south_asian_countries)]

# Select metrics for correlation analysis
metrics = ['Log GDP per capita', 'Social support', 'Healthy life expectancy', 
           'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']

# Calculate correlations with 'Score'
correlations = {metric: south_asia_df['score'].corr(south_asia_df[metric]) for metric in metrics}

# Convert correlations to DataFrame for better readability
correlation_df = pd.DataFrame(correlations.items(), columns=['Metric', 'Correlation']).sort_values(by='Correlation', ascending=False)

# Identify strongest and weakest relationships
strongest = correlation_df.iloc[0]
weakest = correlation_df.iloc[-1]

print("Strongest Relationship:", strongest)
print("Weakest Relationship:", weakest)

# Plot scatter plots for strongest and weakest metrics
plt.figure(figsize=(14, 6))

# Strongest
plt.subplot(1, 2, 1)
sns.regplot(x=strongest['Metric'], y='score', data=south_asia_df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title(f"Strongest Relationship: {strongest['Metric']} vs Score")
plt.xlabel(strongest['Metric'])
plt.ylabel("Score")

# Weakest
plt.subplot(1, 2, 2)
sns.regplot(x=weakest['Metric'], y='score', data=south_asia_df, scatter_kws={'color': 'orange'}, line_kws={'color': 'green'})
plt.title(f"Weakest Relationship: {weakest['Metric']} vs Score")
plt.xlabel(weakest['Metric'])
plt.ylabel("Score")

# Show plots
plt.tight_layout()
plt.show()
```
